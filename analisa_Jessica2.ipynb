```python
#SIAPKAN DATA (AGREGASI KE LEVEL ELEMEN)
df_elemen = (
    df_final
    .groupby(
        ["Umur", "Unit Kerja", "Pendidikan Terakhir", "Elemen"],
        as_index=False
    )["Skor"]
    .mean()
)
```

```python
#ANOVA SKOR ELEMEN BERDASARKAN UMUR
from scipy.stats import f_oneway

print("ANOVA berdasarkan Umur")

for elemen in df_elemen["Elemen"].unique():
    data = [
        group["Skor"].values
        for _, group in df_elemen[df_elemen["Elemen"] == elemen].groupby("Umur")
    ]

    if len(data) > 1:
        f_stat, p_val = f_oneway(*data)
        print(f"{elemen} | F = {f_stat:.3f} | p = {p_val:.4f}")

```

```python
import matplotlib.pyplot as plt

for elemen in df_elemen["Elemen"].unique():
    data = [
        group["Skor"].values
        for _, group in df_elemen[df_elemen["Elemen"] == elemen].groupby("Umur")
    ]
    labels = df_elemen[df_elemen["Elemen"] == elemen]["Umur"].unique()

    plt.figure()
    plt.boxplot(data, labels=labels)
    plt.title(f"Skor Elemen '{elemen}' berdasarkan Umur")
    plt.xlabel("Kelompok Umur")
    plt.ylabel("Skor Elemen")
    plt.show()
```

Ordinary Least Squares. (OLS) Regresi Linier Berganda (Multiple Linear Regression)
dengan Variabel Independen Kategorikal (Dummy Regression)

| Komponen           | Keterangan                                    |
| ------------------ | --------------------------------------------- |
| **Y (Dependen)**   | `Mean_Skor` (rata-rata skor literasi digital) |
| **X (Independen)** | Umur, Unit Kerja, Pendidikan Terakhir         |
| Tipe X             | **Kategorikal** (dummy variables via `C()`)   |

â€œDilakukan analisis regresi linear berganda untuk menguji pengaruh umur, unit kerja, dan pendidikan terakhir terhadap skor literasi digital rata-rata.â€


```python
df_Y = (
    df_final
    .groupby(["Umur", "Unit Kerja", "Pendidikan Terakhir"])
    ["Skor"]
    .mean()
    .reset_index(name="Mean_Skor")
)
```

```python
import statsmodels.formula.api as smf

model = smf.ols(
    "Mean_Skor ~ C(Umur) + C(Q('Unit Kerja')) + C(Q('Pendidikan Terakhir'))",
    data=df_Y
).fit()

print(model.summary())
```

```python
import matplotlib.pyplot as plt

# Ambil koefisien & confidence interval
params = model.params
conf = model.conf_int()
conf.columns = ["Lower", "Upper"]

coef_df = params.to_frame(name="Coef").join(conf)

# Buang intercept agar fokus ke variabel penjelas
coef_df = coef_df.drop("Intercept")

plt.figure(figsize=(8,5))
plt.errorbar(
    coef_df["Coef"],
    coef_df.index,
    xerr=[
        coef_df["Coef"] - coef_df["Lower"],
        coef_df["Upper"] - coef_df["Coef"]
    ],
    fmt='o'
)
plt.axvline(0)
plt.title("Coefficient Plot Regresi Literasi Digital")
plt.xlabel("Nilai Koefisien")
plt.ylabel("Variabel")
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import shapiro

shapiro(model.resid)
```

```python
from statsmodels.stats.diagnostic import het_breuschpagan

bp_test = het_breuschpagan(model.resid, model.model.exog)
bp_test
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

X = model.model.exog
vif = [variance_inflation_factor(X, i) for i in range(X.shape[1])]
vif
```

```python
#Effect Size
def eta_squared(aov):
    aov["eta_sq"] = aov["sum_sq"] / sum(aov["sum_sq"])
    return aov

eta_squared(sm.stats.anova_lm(model, typ=2))
```
