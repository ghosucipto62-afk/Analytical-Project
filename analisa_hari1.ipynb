## Identify Top 10 Customers by Grand Total

### Subtask:
Group the DataFrame by 'ID_Customer' and sum the 'Grand_Total' for each customer. Then, identify and display the top 10 customers with the highest total 'Grand_Total'.


**Reasoning**:
First, I'll group the DataFrame `df` by 'ID_Customer' and sum their 'Grand_Total' to get each customer's total spending. Then, I'll sort these totals in descending order and select the top 10 to identify the most valuable customers.



```python
top_10_customers = df.groupby('ID_Customer')['Grand_Total'].sum().nlargest(10).reset_index()
display(top_10_customers)
```

**Reasoning**:
